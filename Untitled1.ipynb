```python
import random
from IPython.display import display, HTML
import time

# =========================
# DAFTAR PILIHAN
# =========================
items = [
    "AAA",
    "BBB",
    "CCC",
    "DDD",
    "XXX"
]

# =========================
# HASIL PERTAMA DISETTING
# =========================
first_result = "XXX"

# penanda spin pertama
first_spin = True

def spin_wheel():
    global first_spin

    print("🎡 Spinning...")

    for i in range(15):
        print(random.choice(items), end="\r")
        time.sleep(0.1)

    # spin pertama pasti XXX
    if first_spin:
        result = first_result
        first_spin = False
    else:
        result = random.choice(items)

    print(f"\n✅ Hasil: {result}")
    return result

# =========================
# JALANKAN
# =========================

while True:
    spin_wheel()

    again = input("\nSpin lagi? (y/n): ")
    if again.lower() != "y":
        print("Selesai 🎉")
        break
```
