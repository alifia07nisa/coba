```python
#Muat dan Eksplorasi Data

import pandas as pd

url = "https://drive.google.com/uc?id=1wmVjvkNW0fRk3htK0elhaC1c1I9DRKDT"
df = pd.read_csv(url)

print(df.shape)
print(df["Churn"].value_counts(normalize=True))

df.head()
```

```python
#Preprocessing
from sklearn.model_selection import train_test_split
#TODO: encoding fitur kategorikal (mis. pd.get_dummies)
df = pd.get_dummies(df, drop_first=True)

#TODO: pisahkan X (fitur) dan y (target = Churn)
X = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']

#Membagi data menjadi data latih dan data uji
X_tr, X_te, y_tr, y_te = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Ukuran data latih:", X_tr.shape)
print("Ukuran data uji:", X_te.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

#Membangun model Random Forest
rf = RandomForestClassifier(
    n_estimators=300,
    class_weight="balanced",
    random_state=42
)

#Melatih model menggunakan data latih
rf.fit(X_tr, y_tr)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

#Menghitung prediksi
y_pred = rf.predict(X_te)

#Menghitung probabilitas kelas 1 (Churn = Yes)
y_prob = rf.predict_proba(X_te)[:, 1]

#Menampilkan classification report
print(classification_report(y_te, y_pred))

#Menghitung dan menampilkan ROC-AUC
roc_auc = roc_auc_score(y_te, y_prob)
print("ROC-AUC:", roc_auc)
```

```python
#Menghitung probabilitas churn (kelas 1)
churn_prob = rf.predict_proba(X_te)[:, 1]

#Menampilkan 10 probabilitas pertama
#print(churn_prob[:10])

import pandas as pd

hasil = pd.DataFrame({
    "Probabilitas_Churn": churn_prob,
    "Persentase_Churn (%)": (churn_prob * 100).round(2)
})

hasil.head(10)
```

Berdasarkan hasil prediksi probabilitas, Random Forest mampu memberikan estimasi kemungkinan setiap pelanggan akan melakukan churn. Nilai probabilitas yang mendekati 1 atau 100% menunjukkan pelanggan memiliki kemungkinan churn yang lebih tinggi, sedangkan nilai yang mendekati 0 menunjukkan pelanggan memiliki kemungkinan lebih tinggi untuk tetap menggunakan layanan. Hasil prediksi ini dapat digunakan oleh perusahaan untuk mengidentifikasi pelanggan mana saja yang terdapat kemungkinan untuk melakukan churn dan merencanakan strategi agar pelanggan tetap menggunakan layanan.
